```python
# xlwt读取、生成xlwt
import xlwt

workbook = xlwt.Workbook("utf-8")
# 创建sheet
worksheet = workbook.add_sheet("sheet1")
# 写入数据，表单看作二维矩阵，指定行、列、内容
worksheet.write(1,2,"hello world!!!")
workbook.save("demo1.xlsx")
```

```python
# xlrd只能读取数据
from xlrd import open_workbook
wb = open_workbook('D:\C++\data_structure\demo1.xlsx')
sheet1 = wb.sheet_by_name("top250")
rows = sheet1.nrows
cols = sheet1.ncols
# print(rows, cols)
for row in range(rows-1):
    for col in range(cols):
        print(sheet1.cell(row+1, col).value)
```
